```python
import numpy as np
```

```python
file=np.load("projection_dataset.npz")
file.files
```

```python
file["X"].shape, file["y"].shape
```

```python
X_train=file["X"][:2400]
y_train=file["y"][:2400]

X_test=file["X"][2400:]
y_test=file["y"][2400:]

from training_util import train_model
from training_util import PersonaHobbyDataset
from vector_projection_model import HobbyProjector

import torch
from torch.utils.data import Dataset, DataLoader

dataset_train=PersonaHobbyDataset(X_train,y_train)
dataset_test=PersonaHobbyDataset(X_test,y_test)

train_loader = DataLoader(dataset_train, batch_size=128, shuffle=True)
val_loader = DataLoader(dataset_test, batch_size=64, shuffle=False)

model = HobbyProjector(input_dim=X_train.shape[1], output_dim=y_train.shape[1], hidden_dim=4096)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
train_model(model, train_loader, val_loader, num_epochs=100, device=str(device))  
```
